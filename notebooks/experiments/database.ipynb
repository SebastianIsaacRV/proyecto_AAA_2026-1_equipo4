# Database

```python
import duckdb
```

```python
from incidencia_delictiva.config import DATABASE_PATH
```

La conexión: 

```python
con = duckdb.connect(DATABASE_PATH)
```

Las tablas: 

```python
print(con.execute("SHOW TABLES").fetchall())
```

Las tablas por esquema: 

```python
q = """
SELECT table_schema, table_name 
FROM information_schema.tables
"""
db_schema = con.execute(q).fetchdf()
print(db_schema)
```

```python
for tn in db_schema['table_name']:
    # print(con.execute(f"SELECT * FROM {tn} LIMIT 5").fetchdf())
    print(f'Table: {tn}')
    print(con.execute(f"SELECT * FROM {tn} LIMIT 5").fetchdf().dtypes, '\n') 
    # print(con.execute(f"SELECT COUNT(*) FROM {tn}").fetchone())
```

De la tabla `delitos` agrupamos por `año`, `cvegeo`  y agregamos el total de delitos: 

```python
q = """
SELECT 
    año, 
    cvegeo, 
    SUM(total_anual) AS total_delitos
FROM delitos
GROUP BY año, cvegeo
LIMIT 5
"""
con.execute(q).fetchdf()
```
